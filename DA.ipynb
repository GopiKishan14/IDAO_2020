```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(samay, x, test_size=0.9, random_state=42)
```

```python


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout , Activation , Conv2D , MaxPooling2D,Flatten, LSTM
# from keras.utils import plot_model
# from tensorflow.keras.callbacks import TensorBoard

model = Sequential()
model.add(Dense(8, activation=None, input_dim=1))
# model.add(LSTM(16, activation= "tanh"))
model.add(Dense(32, activation="tanh"))
# # Add another:
model.add(Dense(16, activation="tanh"))
# # Add a softmax layer with 10 output units:
model.add(Dense(8, activation="tanh"))

model.add(Dense(1, activation="tanh"))
```

```python
# Configure a model for mean-squared error regression.
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss='mse',       # mean squared error
              metrics=['mae'])  # mean absolute error
```

```python
# model.summary()
# X_train.shape
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=256)
```
