```python
temp = df[df.sat_id==1]
```

```python
# from dateutil.parser import *
import dateutil.parser as dp
# t = '1984-06-02T19:05:00.000Z'

time = list()
t = temp.epoch.iloc[0]
parsed_t = dp.parse(t)
base_time = parsed_t.strftime('%s')

for i in range(len(temp.epoch)):
    time.append(int(dp.parse(temp.epoch.iloc[i]).strftime("%s")) - int(base_time))
    
```

```python
x = {}
```

```python

x[1] = np.array(temp.x)
x[1] = np.expand_dims(x[1] , axis=1)
```

```python
x[1].shape
```

```python
samay  = {}
samay["time"] = np.array(time)/(3600)
samay["x_sim"] = temp.x_sim

x_df = pd.DataFrame(samay)
# samay = np.expand_dims(samay, axis=1)
# samay = np.expand_dims(samay, axis=1)
x_ar = x_df.to_numpy()


x_ar = np.expand_dims(x_ar, axis=2)
x_ar.shape
```

```python
plt.plot(df[df.sat_id==1].x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_ar, x[1], test_size=0.1, random_state=42)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense , Dropout , Activation , Conv2D , MaxPooling2D,Flatten, LSTM
# from keras.utils import plot_model
# from tensorflow.keras.callbacks import TensorBoard
from keras import backend as K

def a(x):
    x = K.sin(x)
    return x

model = Sequential()
# model.add(Dense(8, activation=None, input_dim=1))
model.add(LSTM(1000, activation= "tanh", input_shape=(2,1, ), return_sequences=True))
# model.add(Dense(32, activation="tanh"))
# # Add another:
# model.add(Dense(16, activation="tanh"))
# # Add a softmax layer with 10 output units:
# model.add(Dense(8, activation="tanh"))
model.add(LSTM(1000, activation= a, return_sequences=False))
model.add(Dense(256, activation="tanh"))
# # Add another:
model.add(Dense(32, activation="tanh"))
# # Add a softmax layer with 10 output units:
model.add(Dense(8, activation=a))
model.add(Dense(1, activation="tanh"))
```

```python
# Configure a model for mean-squared error regression.
model.compile(optimizer= keras.optimizers.Adam(0.01), loss='mse', metrics=['mae'])  # mean absolute error
```

```python
model.summary()
# X_train.shape
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=10)
```
